```python
import pandas as pd

dados_educ_pop = pd.read_csv('../dados/dados_educ_pop.csv', sep = ',')
distritos = pd.read_csv('../dados/distritos.csv', sep = ',')
```

```python
dados_gerais = dados_educ_pop.merge(distritos, on='DISTRITO_F')

dist = dados_educ_pop['DISTRITO_F'].unique()

dados_atendimento_2006 = dados_gerais[dados_gerais['MES_ANO_REF'] == 'jun-06'][['DISTRITO_F', 'ZONA', 'MAT_CRECHE', 'TOTAL']]
dados_atendimento_2017 = dados_gerais[dados_gerais['MES_ANO_REF'] == 'dez-17'][['DISTRITO_F', 'ZONA', 'MAT_CRECHE', 'TOTAL']]

dados_atendimento_2006['MAT_TOTAL'] = dados_atendimento_2006['MAT_CRECHE']/dados_atendimento_2006['TOTAL']
dados_atendimento_2017['MAT_TOTAL'] = dados_atendimento_2017['MAT_CRECHE']/dados_atendimento_2017['TOTAL']

dados_atendimento_2006 = dados_atendimento_2006.sort_values(by='MAT_TOTAL', ascending=False).reset_index(drop=True)
dados_atendimento_2017 = dados_atendimento_2017.sort_values(by='MAT_TOTAL', ascending=False).reset_index(drop=True)

for index in dados_atendimento_2017.index:
    nome = dados_atendimento_2017.at[index, 'DISTRITO_F']
    dados = dados_atendimento_2006[dados_atendimento_2006['DISTRITO_F'] == nome]
    caracter = '\mesmo'
    diferenca = ''
    zona = dados_atendimento_2017.at[index, 'ZONA']
    ranking = dados.index[0]+1
    ranking_novo = index+1
    if (ranking_novo < ranking):
        caracter = '\\aumento'
        diferenca = abs(ranking_novo - ranking)
    if (ranking_novo > ranking):
        caracter = '\queda'
        diferenca = abs(ranking_novo - ranking)
    print("\\textbf{" + str(ranking_novo) + "} & " + caracter + ' ' + str(diferenca) + " & " + nome + " & " + zona +  " & " + str(round(dados_atendimento_2017.at[index, 'MAT_TOTAL']*100, 2)) + "\\\\ \hline")
```
